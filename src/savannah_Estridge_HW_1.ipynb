```python
#Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Import MTCARS data using pd.read_csv()


```python
# Import mtcars data and define as a dataframe

path = r"https://vincentarelbundock.github.io/Rdatasets/csv/datasets/mtcars.csv"
cars = pd.read_csv(path)
```

# 2. Display the first five rows of cars


```python
# Display first five rows
display(cars.head(5))
```

# 3. For each category in the cyl column, calculate the average mpg for all cars with that cyl value.

```python
# For reference https://www.geeksforgeeks.org/pandas/python-pandas-dataframe-groupby/

cars.groupby('cyl')['mpg'].mean()
```

# 4. Create a histogram using the mpg column


```python
# Define set of bins
bins = [5, 10, 15, 20, 25, 30, 35]

# Define Plot
sns.set_style("whitegrid")
sns.set_context("talk", font_scale= .8)
ax = sns.histplot(cars, x = "mpg", bins = bins)

# Format axis and title
ax.set(
    title="Miles Per Gallon (MPG) Histogram",
    xlabel="MPG",
    ylabel="Count"
)

plt.show()
```

# 5. Choose two variables in the data and create a scatterplot.

```python
sns.set_style("whitegrid")
sns.set_context("talk", font_scale= .8)
ax = sns.scatterplot(cars, x = "wt", y = "mpg", hue = "hp")
ax.set(
    title= "Miles Per Gallon (MPG) VS. Weight",
    xlabel= "Weight",
    ylabel= "MPG"
)

plt.show()
```

# 1. [Social Media Addication and Usage Pattern Dataset:](https://www.kaggle.com/datasets/abdullahmeo/social-media-addiction-and-usage-patterns-dataset?select=social_media_addiction.csv)

This [kaggle dataset](https://www.kaggle.com/) analyzes individual social media usage behavior, focusing on engagement intensity, addiction levels, emotional impact, productivity loss, and mental health indicators.  

```python
from pathlib import Path
import kagglehub
import pandas as pd


# DOwnload and define path of kaggle dataset
path = Path(kagglehub.dataset_download(
    "abdullahmeo/social-media-addiction-and-usage-patterns-dataset"
))
files = list(path.iterdir())
csv_file = next(path.glob("*.csv"))

#Read CSV as a DF
df = pd.read_csv(csv_file)
```

# 2. Display first 5 rows

```python
# Display the first 5 rows
display(df.head(5))
```

# 3. Create a visualization using two variables from this data.

```python
# Define Plot
sns.set_style("whitegrid")
sns.set_context("talk", font_scale= .8)
ax = sns.scatterplot(df, x= "Age", y= "Daily_Usage_Time_min", hue = "Gender")

# Format axis and title
ax.set(
    title="Age VS. Daily Usage Time",
    xlabel="Age",
    ylabel="Minutes"
)

plt.show()
```
